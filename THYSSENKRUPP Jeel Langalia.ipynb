```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import LabelEncoder

# Define features (X) and target variable (y)
X_mathematics = df1.drop(['G3'], axis=1)  # Features
y_mathematics = df1['G3']  # Multiclass classification based on 'G3' (pass or fail)

# Encode categorical variables (if needed)
label_encoder = LabelEncoder()
for column in X_mathematics.select_dtypes(include='object').columns:
    X_mathematics[column] = label_encoder.fit_transform(X_mathematics[column])

# Split the data into training and testing sets
X_train_mathematics, X_test_mathematics, y_train_mathematics, y_test_mathematics = train_test_split(X_mathematics, y_mathematics, test_size=0.2, random_state=42)

# Train a Logistic Regression model
classifier_mathematics = LogisticRegression(random_state=42)
classifier_mathematics.fit(X_train_mathematics, y_train_mathematics)

# Make predictions on the test set
predictions_mathematics = classifier_mathematics.predict(X_test_mathematics)

# Evaluate the model
accuracy_mathematics = accuracy_score(y_test_mathematics, predictions_mathematics)
print("Accuracy:", accuracy_mathematics)
print("\nClassification Report:\n", classification_report(y_test_mathematics,predictions_mathematics))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import LabelEncoder

# Define features (X) and target variable (y) for regression
X_regression_mathematics = df1.drop(['G3'], axis=1)  # Features
y_regression_mathematics = df1['G3']  # Regression task predicting 'G3' values (0 to 20)

# Encode categorical variables (if needed)
label_encoder = LabelEncoder()
for column in X_regression_mathematics.select_dtypes(include='object').columns:
    X_regression_mathematics[column] = label_encoder.fit_transform(X_regression_mathematics[column])

# Split the data into training and testing sets
X_train_regression_mathematics, X_test_regression_mathematics, y_train_regression_mathematics, y_test_regression_mathematics = train_test_split(
    X_regression_mathematics, y_regression_mathematics, test_size=0.2, random_state=42
)

# Train a Linear Regression model
regressor_regression_mathematics = LinearRegression()
regressor_regression_mathematics.fit(X_train_regression_mathematics, y_train_regression_mathematics)

# Make predictions on the test set
predictions_regression_mathematics = regressor_regression_mathematics.predict(X_test_regression_mathematics)

# Evaluate the model
mse_regression_mathematics = mean_squared_error(y_test_regression_mathematics, predictions_regression_mathematics)
r2_regression_mathematics = r2_score(y_test_regression_mathematics, predictions_regression_mathematics)

print("Mean Squared Error (Regression):", mse_regression_mathematics)
print("R-squared Score (Regression):", r2_regression_mathematics)

# Visualize the actual vs predicted values for regression
plt.figure(figsize=(10, 6))
sns.scatterplot(x=y_test_regression_mathematics, y=predictions_regression_mathematics)
plt.title('Actual vs Predicted Values for G3 (Regression)')
plt.xlabel('Actual G3 Values')
plt.ylabel('Predicted G3 Values')
plt.show()
```
